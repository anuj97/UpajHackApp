```python
import pandas as pd
data = pd.read_csv('rainfall_data.csv')
data.head()
```

```python
data = data.loc[data['SUBDIVISION'] == 'jammu and kashmir']
```

```python
data = data.fillna(data.mean())
```

```python
year = []
rain = []
for r in data['ANNUAL']:
    rain.append(r)
for y in data['YEAR']:
    year.append(str(y))
print(year)
print(rain)
```

```python
data
```

```python
df = pd.DataFrame({'rainfall' : rain, 'years' : year} , index = year)
plot = df.plot.line(y='rainfall',x = 'years',  figsize=(5,5), grid = True)
```
