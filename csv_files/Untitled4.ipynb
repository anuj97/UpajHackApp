```python
import pandas as pd
data = pd.read_csv('check.csv')
data.head()
```

```python
data = data.groupby(['state', 'crop'], as_index=False)['production'].sum()
```

```python
data.head()
```

```python
data = data.loc[data['state'] == 'uttar pradesh']
data.head()
```

```python
data = data.groupby('crop', as_index=False)['production_by_area'].sum()
```

```python
data
```

```python
data = data.head()
crop = []
prod = []
for obj in data['crop']:
    crop.append(obj)
for obj in data['production_by_area']:
    prod.append(obj)
```

```python
df = pd.DataFrame({'production' : prod}, index = crop)
```

```python
plot = df.plot.pie(y='production', figsize=(5,5))
```
